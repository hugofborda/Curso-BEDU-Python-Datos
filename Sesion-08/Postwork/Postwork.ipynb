# Filtro de datos tasa de conversión de dólares a pesos
Se va a realizar el filtro de datos de tasa de conversión con el fin de dejar la granularidad a nivel de año

Obtener datos transformados de tasa de conversión

```python
tasa_conversion = pd.read_csv("../../DataSets/Postwork/TasaConversion.csv",  index_col= 0)
tasa_conversion['Fecha'] = pd.to_datetime(tasa_conversion['Fecha'],format='%Y-%m-%d')
```

Obtener la tasa de conversión obtenida al final de cada año

```python
tasa_conversion = tasa_conversion[ (pd.DatetimeIndex(tasa_conversion["Fecha"]).month == 12) & (pd.DatetimeIndex(tasa_conversion["Fecha"]).day == 31) ]
tasa_conversion = tasa_conversion.drop(columns=["Fecha"]).set_index("Anio", drop=True)
```

Obtener tasa_anterior, para calcular variacion de un año a otro

```python
tasa_conversion = tasa_conversion.sort_values("Anio")
tasa_conversion_ant = tasa_conversion.copy()
tasa_conversion_ant = tasa_conversion_ant.reset_index()
```

Generar registro año anterior, para realizar el merge con el año del periodo a evaluar

```python
tasa_conversion_ant["AnioAnterior"] = tasa_conversion_ant["Anio"] - 1
tasa_conversion_ant = tasa_conversion_ant.set_index("AnioAnterior", drop = True)
```

Generar dataframe con población final de tasa de conversión con variacion

```python
tasa_variacion = pd.merge(tasa_conversion, tasa_conversion_ant , left_on="Anio", right_index=True)
tasa_variacion["Variacion"] = ( tasa_variacion["TRM_y"] - tasa_variacion["TRM_x"]  )  / tasa_variacion["TRM_y"] * 100
tasa_variacion = tasa_variacion[ ["Anio","TRM_y","Variacion"] ]
tasa_variacion = tasa_variacion.rename(columns={"TRM_y" : "TRM"})
tasa_variacion = tasa_variacion.set_index("Anio", drop = True)
```

almacenar resultado en archivo CSV

```python
tasa_variacion.to_csv("../../DataSets/Postwork/TasaConversionAnual.csv")
```

# Agrupación de datos y Dataframe final
Realizar el merge de los conjuntos resultantes

Obtener datos de archivos CSV

```python
tasa_conversion_anual = pd.read_csv("../../DataSets/Postwork/TasaConversionAnual.csv",  index_col= 0)
salario_minimo_anual = pd.read_csv("../../DataSets/Postwork/SalarioMinimoAnual.csv",  index_col= 0)
inflacion_anual = pd.read_csv("../../DataSets/Postwork/InflacionAnual.csv",  index_col= 0)
```

Generar dataframe final

```python
tasa_salario = pd.merge(tasa_conversion_anual, salario_minimo_anual, left_on="Anio", right_index=True)
tasa_salario = tasa_salario.rename(columns={"Variacion_x": "Variacion_TRM", "Variacion_y": "Variacion_Salario"})
tasa_salario_inflacion = pd.merge(tasa_salario, inflacion_anual, left_on="Anio", right_index=True)
tasa_salario_inflacion = tasa_salario_inflacion.rename(columns={"InflacionAnual": "Variacion_Inflacion"})
tasa_salario_inflacion["Variacion_TRM"] = round(tasa_salario_inflacion["Variacion_TRM"],2)
tasa_salario_inflacion["Variacion_Salario"] = round(tasa_salario_inflacion["Variacion_Salario"],2)
tasa_salario_inflacion["Variacion_Inflacion"] = round(tasa_salario_inflacion["Variacion_Inflacion"],2)
```

Almacenar dataframe final

```python
tasa_salario_inflacion.to_csv("../../DataSets/Postwork/TasaSalarioInflacionAnual.csv")
```
