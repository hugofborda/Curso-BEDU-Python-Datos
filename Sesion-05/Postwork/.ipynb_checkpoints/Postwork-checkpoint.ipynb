# Trasnformación de datos tasa de conversión de dólares a pesos

```python
import pandas as pd
```

```python
dfTasaConversionDP = pd.read_excel('../../DataSets/Postwork/1.1.1.TCM_Serie historica IQY.xlsx',
                 skiprows = 7,)
dfTasaConversionDP
```

```python
dfTasaConversionDP.isna().sum()
```

```python
dfTasaConversionDP.isna().sum(axis=1)
```

```python
dfTasaConversionDP_1 = dfTasaConversionDP.dropna(axis=0,how='any')
dfTasaConversionDP_1.tail()
```

```python
nombres_tasa = {
    'Fecha (dd/mm/aaaa)': 'Fecha',
    'Tasa de cambio representativa del mercado (TRM)': 'TRM',
}
dfTasaConversionDP_1 = dfTasaConversionDP_1.rename(columns=nombres_tasa)
dfTasaConversionDP_1 = dfTasaConversionDP_1.reset_index(drop=True)
dfTasaConversionDP_1
```

# Transfomación datos inflación

```python
dfInflacion = pd.read_excel('../../DataSets/Postwork/1.2.5.IPC_Serie_variaciones.xlsx',
                 skiprows = 12,)
dfInflacion
```

```python
dfInflacion.isna().sum()
```

```python
dfInflacion.isna().sum(axis=1)
```

```python
dfInflacion_1 = dfInflacion.dropna(axis=0,how='any')
dfInflacion_1.tail()
```

```python
dfInflacion_1.isna().sum()
```

```python
nombres_inflacion = {
    'Año(aaaa)-Mes(mm)': 'AnioMes',
    'Índice': 'Indice',
    'Inflación anual %': 'InflacionAnual',
    'Inflación mensual %': 'InflacionMensual',
    'Inflación año corrido %': 'InflacionAnualCorrida',
    
}
dfInflacion_1 = dfInflacion_1.rename(columns=nombres_inflacion)
dfInflacion_1 = dfInflacion_1.reset_index(drop=True)
dfInflacion_1
```
