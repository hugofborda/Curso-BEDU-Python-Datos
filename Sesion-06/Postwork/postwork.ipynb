# API COVID Estados Unidos

```python
import pandas as pd
import requests
```

```python
url = 'https://api.covidtracking.com/v1/states/'
urlMethod = 'current.json'
```

```python
response = requests.get(url + urlMethod)
```

```python
response.status_code
```

```python
datos = response.json()
datos
```

```python
dfCurrent = pd.json_normalize(datos)
dfCurrent
```

```python
urlMethod2 = 'daily.json'
```

```python
response2 = requests.get(url + urlMethod2)
```

```python
dfDaily = pd.json_normalize(response2.json())
dfDaily
```
